# lags, changes, percentage changes

```python
df_22_open = df_22[['Open']]
```

```python
df_22_open.head()
```

```python
# one period forward in time
df_22_open['Shift'] = df_22_open.Open.shift() # default periods = 1
df_22_open
```

```python
df_22_open['Lagged'] = df_22_open['Open'].shift(periods=-1)
df_22_open = df_22_open[['Open', 'Lagged','Shift' ]]
df_22_open
```

```python
# calculate one period percentage change
df_22_open['Changed'] = df_22_open['Open'].div(df_22_open['Shift'])
```

```python
# now subtract by 1 and multiply by 100 get the relative change in percentage terms

df_22_open['Change'] = df_22_open['Changed'].sub(1).mul(100).round(2)
```

```python
df_22_open
```

```python
# can also use diff() which is returns the difference in values for to adjacent time periods

df_22_open['Diff'] = df_22_open['Open'].diff()
```

```python
df_22_open
```

```python
15.695195-15.689439
```

```python
# built in pandas method to calcualte percent change directly

df_22_open['Pct_Change'] = df_22_open['Open'].pct_change().mul(100).round(2)
```

```python
df_23 = pd.read_csv('Google_Stock_Test (2023).csv')
df_23['Date'] = pd.to_datetime(df_23['Date'])
df_23
```

```python
df= pd.read_csv('Google_Stock_Train (2010-2022).csv')
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
# stacking both dfs
data = pd.concat([df, df_23], ignore_index=True)
data
```

# Now compare Avg 'Open' percentage change YoY

```python
# now compare montly 'Open' percentage change YoY

data['year'] = data['Date'].dt.year
data['month'] = data['Date'].dt.month
```

```python
data
```

```python
data.groupby(['year'])['Open'].mean().reset_index()
```

```python
annual_avg = data.groupby(['year'])['Open'].mean().reset_index()
annual_avg = annual_avg.rename(columns={'Open': 'annual_avg_open'})
annual_avg.head()
```

```python
# Calculate avg open price YoY perc change 

annual_avg['pct_change'] = annual_avg['annual_avg_open'].pct_change().round(2)
```
